```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sal_train = pd.read_csv("SalaryData_Train(1).csv")
sal_test =pd.read_csv("SalaryData_Test(1).csv")
```

```python
sal_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoding = LabelEncoder()
```

```python
sal_train['Salary']=encoding.fit_transform(sal_train.Salary)
```

```python
sal_test['Salary']=encoding.fit_transform(sal_test.Salary)
```

```python
sal_train
```

```python
plt.scatter(sal_train['age'],sal_train['Salary'])
```

```python
sal_train = pd.get_dummies(sal_train)
```

```python
sal_train
```

```python
x_train = sal_train.drop(columns=['Salary'])
```

```python
y_train =sal_train['Salary']
```

```python
sal_test =pd.get_dummies(sal_test)
```

```python
sal_test
```

```python
x_test = sal_test.drop(columns=['Salary'])
```

```python
y_test = sal_test['Salary']
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```
